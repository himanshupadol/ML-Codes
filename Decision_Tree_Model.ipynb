## Importing necessary libraries

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import load_breast_cancer
```

## Loading the dataset

```python
data = load_breast_cancer()
```

## Creating a dataframe of the dataset

```python
print(data)
```

```python
updata = pd.DataFrame(data.data, columns=data.feature_names)
```

## High Level Summary of the dataset

```python
updata.head()
```

```python
updata.shape
```

```python
updata['Label'] = data.target
updata.head()
```

```python
updata.shape
```

```python
updata['Label'].value_counts()
```

```python
updata.info()
```

## Splitting the dataset in training and testing subsets

```python
x=updata.iloc[:, :30]
# or we can directly use x=data.data & y=data.target
x.head()
```

```python
y=updata.iloc[:, 30]
y.head()
```

Here, x & y are predictors and target features respectively.

#### **Creating training and testing subsets**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=4)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

## Training the decision tree model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treemodel = DecisionTreeClassifier(criterion='gini')
```

```python
treemodel.fit(x_train, y_train)
```

```python
trainpred = treemodel.predict(x_train)
```

#### Checking model's performance on training subset

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_train, trainpred)
```

```python
pd.crosstab(y_train, trainpred)
```

## Testing the decision tree model

```python
testpred = treemodel.predict(x_test)
```

#### Checking model' performance on testing subset

```python
accuracy_score(y_test, testpred)
```

```python
pd.crosstab(y_test, testpred)
```

**NOTE**: The decrease in accuracy represents overfitting.

# Creating the classification program

```python
# Variable for new input data
InputData = (9.504, 12.44, 60.34, 273.9, 0.1024, 0.06492, 0.02956, 0.02076, 0.1815, 0.06905, 0.2773, 0.9768, 1.909, 15.7, 0.009606, 0.01432, 0.01985, 0.01421, 0.02027, 0.002968, 10.23, 15.66, 65.13, 314.9, 0.1324, 0.1148, 0.08867, 0.06227, 0.245, 0.07773)

# Converting new input data into array
InputArry = np.asarray(InputData)

# Reshaping
reshaped = InputArry.reshape(1,-1)

# Classification Task
prog_pred = treemodel.predict(reshaped)
print("The input data of breast cancer test outputs", prog_pred)

if prog_pred==1:
  print("The tested tumor is not cancerous or we can say it is harmless.")
else:
  print("The tested tumor is of serious concern and has probability of growing uncontrollably.")
```

# Visualizing the prediction activity

```python
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 8))
tree.plot_tree(treemodel, feature_names=data.feature_names, class_names=data.target_names, filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```

### Alternate Method

```python
tree.export_graphviz(treemodel, out_file="tree.dot", feature_names=data.feature_names, class_names=data.target_names, filled=True)
```
